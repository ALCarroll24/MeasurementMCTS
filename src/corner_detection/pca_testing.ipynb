```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import least_squares
import timeit
from sklearn.decomposition import PCA
```

```python
# Load the pickle files from data folder
points1 = np.load(f'data/3d_points_0.pkl', allow_pickle=True)
# points2 = np.load(f'data/points_2.pkl', allow_pickle=True)
# points3 = np.load(f'data/points_3.pkl', allow_pickle=True)

points1_numpy = np.zeros((len(points1), 3))
for i, tuple in enumerate(points1):
    points1_numpy[i,:] = np.array([tuple[0], tuple[1], tuple[2]])
    
# plot the points in 2d
fig, ax = plt.subplots()
ax.scatter(points1_numpy[:,0], points1_numpy[:,1])
```

```python
z_mean = np.mean(points1_numpy[:,2])
z_std = np.std(points1_numpy[:,2])

print(z_mean, z_std)
print(f'Z highest: {np.max(points1_numpy[:,2])}')
print(f'Z lowest: {np.min(points1_numpy[:,2])}')

fig, ax = plt.subplots()
ax.scatter(points1_numpy[:,0], points1_numpy[:,1])

# Remove outliers
points1_numpy = points1_numpy[points1_numpy[:,2] < z_mean + 0*z_std]

pca = PCA(n_components=2)
transformed_points = pca.fit_transform(points1_numpy[:,:2])

# plot the points in 2d
fig, ax = plt.subplots()
ax.scatter(transformed_points[:,0], transformed_points[:,1])
```
